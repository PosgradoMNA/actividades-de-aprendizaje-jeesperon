```python
import numpy as np

df_cars = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
import pandas as pd
df_carsfile = pd.read_csv(df_cars)
df_carsfile.head(5)
```

Check the bottom 10 rows of data frame "df".

```python
df_carsfile.tail(10)
```

```python
# create headers list
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df_carsfile.columns = headers
df_carsfile.head(10)
```

```python
df1=df_carsfile.replace('?',np.NaN)
df=df1.dropna(subset=["price"], axis=0)
df.head(20)
```

```python
print(df.columns)
```

```python
df.to_csv("automobile.csv", index=False)
```
